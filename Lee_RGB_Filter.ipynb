```python
from io import BytesIO
import numpy as np
from PIL import Image
import requests
import matplotlib.pyplot as plt #all of the things that I imported

def red_filter(pic): #all of the filter functions
  for row in pic:
    for pixel in row:
      pixel[1] = 0
      pixel[2] = 0

def green_filter(pic):
  for row in pic:
    for pixel in row:
      pixel[0] = 0
      pixel[2] = 0

def blue_filter(pic):
  for row in pic:
    for pixel in row:
      pixel[0] = 0
      pixel[1] = 0

def redgreen(pic):
  for row in pic:
    for pixel in row:
      pixel[2] = 0

def redblue(pic):
  for row in pic:
    for pixel in row:
      pixel[1] = 0

def greenblue(pic):
  for row in pic:
    for pixel in row:
      pixel[0] = 0

def show_pic(images, title1): #the acutal function which prints the images and the title
  plt.imshow(images)
  plt.axis('off')
  plt.title(title1)
  plt.show()
  return

print("Please give a jpg image")
userinput = input()
response = requests.get(userinput) #stores the response
image = np.array(Image.open(BytesIO(response.content))) #converts the image into a Numpy array
copy_image = np.copy(image) #multiple copies for the filters
copy_image1 = np.copy(image)
copy_image2 = np.copy(image)
copy_image3 = np.copy(image)
copy_image4 = np.copy(image)
copy_image5 = np.copy(image)

show_pic(image, 'Original') #calls of the function to print and filter
red_filter(copy_image)
show_pic(copy_image, 'Red filtered')
green_filter(copy_image1)
show_pic(copy_image1, 'Green filtered')
blue_filter(copy_image2)
show_pic(copy_image2, 'Blue filtered')
redgreen(copy_image3)
show_pic(copy_image3, 'Red and Green')
redblue(copy_image4)
show_pic(copy_image4, 'Red and Blue')
greenblue(copy_image5)
show_pic(copy_image5, 'Green and Blue')
combined_image = np.add(copy_image, copy_image1)
combined_image2 = np.add(combined_image, copy_image2)
equal1 = np.array_equal(image, combined_image2)
if equal1 == True:
  print('this stuff matches')
```
