1.  Load the following dataset into Colab and perform the requested operations in code.

 Source: https://ds1002-resources.s3.amazonaws.com/data/ride_sharing.csv

```python
# Load pandas and the data
import pandas as pd
rides = pd.read_csv('./ride_sharing.csv')
```

```python
# show how many rows and columns are in this dataset
df = pd.DataFrame(rides) #assigning the dataset to a dataframe
df
```

```python
# The first column, "duration" is seen as a string.
# Using two snippets of code, clean this column up so that
# is now an integer (int64)
df['duration'] = df['duration'].str.replace('minutes', '')
df['duration'] = df['duration'].str.replace('minute', '')
df['duration'] = df['duration'].astype(int) #removing the minutes and minute from the duration column so it can be converted to an integer
df
```

```python
# Now display the datatypes in the updated dataframe
df_types = df.dtypes #finding out the types in the new dataframe
df_types
```

```python
# Show the min and max values of ride durations (the shortest and longest rides)
# Hint: If 'df' is the name of your dataframe, simply specify the column name
# and then append the .min() or .max() methods.
dura_min = df['duration'].min() #finds the min of duration
dura_max = df['duration'].max() #finds the max of duration
print(dura_min, dura_max)
```

```python
# Create a new dataframe object and populate it with rides that are 60 minutes or less.
# Hint: Use the .query() method for the dataframe that we used in class.
df1 = df.query('duration <= 60') #gives the rows where the duration is less than or equal to 60
df1
```

```python
# Now show both the min and max values of the duration column for your new dataset.
dura_min1 = df1['duration'].min() #checks new min for duration
dura_max1 = df1['duration'].max() #checks new max for duration
print(dura_min1, dura_max1)
```

```python
# Using the new dataframe, update it so that is ONLY has records where
# the tire_size is 26.
df1 = df1.query('tire_size == 26') #same as before gives only tire size of 26
df1
```

```python
# Finally, produce a count of all records with FEMALE riders
# using the 26-inch wheeled rental bikes for their rentals of
# one hour or less.
df2 = df.query('duration <= 60 and user_gender == "Female" and tire_size == 26') #gives rows with females, duration of less than or equal to 60 and tire size of 26
df_leng = len(df2)
df_leng
```

The `re` package has various functions related to regular expressions, including for example `match()`, which takes two arguments, first the regular expression and second the test string, and tests to see whether the *beginning* of the `test` string matches the pattern `exp`.
